# This notebook illustrates how to annotate a discrete object of the same type from a fluorescent image  


In this notebook, we will annotate a fluorescent image of a IBEX protein multiplex imaging. 
We will load the image, and then annotate it using the object annotation tool by manually drawing polygons around thymus lobules.
Annotations will be saved with the extension .pkl.

```python
import os
import numpy as np
import pandas as pd
import panel as pn
import socket
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import tissue_tag as tt
host = '8889' # set the port to the value in the address bar
```

## Load image and define annotation names and colors 

```python
# here you can either read a single image (grayscale or RGB) or generate a virtual H&E from 2 images in the next cell
# set path
path_to_tissue_tag = # path to tissue tag main folder (e.g. c:/Users/John/Downloads/TissueTag)
path = path_to_tissue_tag + '/data/tissue_tag_minimal_example_ibex/' 
im,_,ppm = tt.read_image(path+'Sample_05_THY45_Z5_ch0009.jpg',contrast_factor=2,background_image_path=path+'Sample_05_THY45_Z5_ch0058.jpg',ppm_image=3.5198,ppm_out=1)
plt.figure(figsize=(8, 6), dpi=200)
plt.imshow(im,origin='lower')
plt.show()

# define annotations "lobules"  
annodict = {
    'lobes':'red',
}
```

```python
# at this stage, if you are adding annotations make sure they come after the current ones otherwise it will mess up everything
labels = np.zeros((im.shape[0],im.shape[1]), dtype=np.uint8) # blank annotation image
labels_rgb = tt.rgb_from_labels(labels,colors=list(annodict.values()))
out_img = tt.overlay_labels(im,labels_rgb,alpha=0.5,show=False)
tabs,render_dict = tt.poly_annotator(im,out_img,anno_dict=annodict,use_datashader=True)

pn.io.notebook.show_server(tabs, notebook_url=f'{socket.gethostname()}:'+host) # determine if working on a server or local environment, if working on a local pc - 'local', while working on server instance indicate the port number e.g. 5011 (string)
```

## plot and update annotations 

```python
labels,object_dict  = tt.object_annotator(im,labels,anno_dict=annodict,render_dict=render_dict,alpha=0.5)
object_dict
# plot result 
labels_rgb = tt.rgb_from_labels(labels,colors=list(object_dict.values()))
out_img = tt.overlay_labels(im,labels_rgb,alpha=0.5)
```

Save annotations

```python
isExist = os.path.exists(path+'tissue_annotations')
if not(isExist):
    os.mkdir(path+'/tissue_annotations/')
    
tt.save_annotation(
    folder=path+'/tissue_annotations/',
    label_image=labels,
    file_name='annotations_'+list(annodict.keys())[0],
    anno_names=list(object_dict.keys()),
    anno_colors=list(object_dict.values()),
    ppm = ppm,
    
)
```
